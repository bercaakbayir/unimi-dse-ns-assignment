```python
import csv

products_file = "amazon-meta.txt"
output_file = "amazon_products.csv"

# Dictionary to store results
id_to_info = {}

with open(products_file, "r", encoding="utf-8") as f:
    current_id = None
    current_title = None
    current_group = None
    
    for line in f:
        line = line.strip()
        if line.startswith("Id:"):
            # Save previous product if it exists
            if current_id is not None:
                title = current_title if current_title is not None else f"Unknown-{current_id}"
                group = current_group if current_group is not None else "Unknown"
                id_to_info[current_id] = (title, group)
            
            # Start new product
            current_id = int(line.split(":")[1].strip())
            current_title = None
            current_group = None
            
        elif line.startswith("title:"):
            current_title = line.split("title:")[1].strip()
        elif line.startswith("group:"):
            current_group = line.split("group:")[1].strip()

    # Save last product
    if current_id is not None:
        title = current_title if current_title is not None else f"Unknown-{current_id}"
        group = current_group if current_group is not None else "Unknown"
        id_to_info[current_id] = (title, group)

# Write to CSV
with open(output_file, "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["id", "product_name", "group"])  # header
    for pid, (title, group) in id_to_info.items():
        writer.writerow([pid, title, group])

print(f"Dataset saved to {output_file} with {len(id_to_info)} products.")
```
