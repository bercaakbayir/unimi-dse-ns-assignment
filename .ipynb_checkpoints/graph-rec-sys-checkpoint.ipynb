# Recommendation Systems

## Graph-Based Rec Sys

### PageRank Algorithm

This algorithm is appropriate for your co-purchase graph because it efficiently computes a localized, approximate Personalized PageRank (PPR), which is ideal for recommending related products in a large, sparse network. The Andersen-Chung-Lang push method focuses computation only on nodes reachable from the source with significant residuals, avoiding unnecessary computation over the entire graph of 262k nodes and 1.23M edges. Its design naturally handles the graph’s hub-and-spoke structure, where a few high-degree nodes dominate connections but most nodes have low degrees, allowing accurate ranking of the most relevant co-purchased items without exhaustive traversal. Additionally, by controlling the residual threshold epsilon and restart probability alpha, the algorithm balances runtime, memory usage, and recommendation locality, making it highly suitable for generating personalized recommendations in a directed, weakly clustered co-purchase network.

```python
def appr(G, source, alpha=0.85, epsilon=1e-4):
    """
    Approximate Personalized PageRank using Andersen-Chung-Lang push algorithm.
    
    Parameters:
        G: directed graph (NetworkX)
        source: starting node
        alpha: restart probability
        epsilon: accuracy threshold
        
    Returns:
        p: PPR score vector {node: score}
    """
    p = {n: 0.0 for n in G.nodes()}
    r = {n: 0.0 for n in G.nodes()}
    r[source] = 1.0

    queue = deque([source])

    while queue:
        v = queue.popleft()
        rv = r[v]
        
        # Only push when residual is large
        if rv < epsilon * G.out_degree(v):
            continue
        
        # Push probability into p
        p[v] += alpha * rv

        # Distribute residual to neighbors
        push_val = (1 - alpha) * rv / G.out_degree(v) if G.out_degree(v) > 0 else 0
        
        r[v] = 0  # consumed residual

        for nbr in G.successors(v):
            old = r[nbr]
            r[nbr] += push_val
            # If residual is now large enough to push, add to queue
            if old < epsilon * G.out_degree(nbr) <= r[nbr]:
                queue.append(nbr)

    return p


def recommend_appr(G, node, top_k=10, alpha=0.7, epsilon=1e-4):
    ppr_scores = appr(G, node, alpha=alpha, epsilon=epsilon)
    ppr_scores.pop(node, None)  # remove self

    # Get top-k nodes sorted by APPR score
    return heapq.nlargest(top_k, ppr_scores, key=ppr_scores.get)
```

### Evaluation

```python
import math

def precision_at_k(recommended, ground_truth, k):
    rec_k = recommended[:k]
    hits = len(set(rec_k) & set(ground_truth))
    return hits / k

def recall_at_k(recommended, ground_truth, k):
    rec_k = recommended[:k]
    hits = len(set(rec_k) & set(ground_truth))
    return hits / len(ground_truth) if ground_truth else 0

def mrr_at_k(recommended, ground_truth, k):
    for rank, item in enumerate(recommended[:k], start=1):
        if item in ground_truth:
            return 1 / rank
    return 0

def average_precision_at_k(recommended, ground_truth, k):
    hits, score = 0, 0
    for i, item in enumerate(recommended[:k], start=1):
        if item in ground_truth:
            hits += 1
            score += hits / i
    return score / len(ground_truth) if ground_truth else 0

def ndcg_at_k(recommended, ground_truth, k):
    dcg = 0.0
    for i, item in enumerate(recommended[:k], start=1):
        if item in ground_truth:
            dcg += 1 / math.log2(i + 1)
    ideal_hits = min(len(ground_truth), k)
    idcg = sum(1 / math.log2(i + 1) for i in range(1, ideal_hits + 1))
    return dcg / idcg if idcg > 0 else 0
```

```python
def train_test_split_edges(G, test_ratio=0.2):
    G_train = G.copy()
    removed_edges = {}

    import random
    random.seed(42)

    for node in G.nodes():
        succ = list(G.successors(node))
        if len(succ) <= 1:
            continue

        k = max(1, int(len(succ) * test_ratio))
        test_edges = random.sample(succ, k)
        removed_edges[node] = test_edges

        for t in test_edges:
            if G_train.has_edge(node, t):
                G_train.remove_edge(node, t)

    return G_train, removed_edges
```

```python
def recommend(G, node, method='neighbors', top_k=10):
    if method == 'neighbors':
        return recommend_neighbors(G, node, top_k)
    elif method == 'appr':
        return recommend_appr(G, node, top_k)
    else:
        raise ValueError("Method must be 'neighbors', 'pagerank', or 'appr'")
```

```python
def evaluate_node(node, G, removed_edges, method='neighbors', top_k=10):
    ground_truth = removed_edges.get(node, [])
    if not ground_truth:
        return None
    
    recommended = recommend(G, node, method, top_k)

    return {
        'precision': precision_at_k(recommended, ground_truth, top_k),
        'recall': recall_at_k(recommended, ground_truth, top_k),
        'mrr': mrr_at_k(recommended, ground_truth, top_k),
        'map': average_precision_at_k(recommended, ground_truth, top_k),
        'ndcg': ndcg_at_k(recommended, ground_truth, top_k),
    }
```

```python
from concurrent.futures import ThreadPoolExecutor, as_completed
import numpy as np

def evaluate_graph_recommendations_parallel(G, removed_edges, target_nodes, method='neighbors', top_k=10, n_jobs=4):
    metrics = {m: [] for m in ['precision', 'recall', 'mrr', 'map', 'ndcg']}

    with ThreadPoolExecutor(max_workers=n_jobs) as executor:
        futures = {
            executor.submit(evaluate_node, node, G, removed_edges, method, top_k): node
            for node in target_nodes
        }

        for future in as_completed(futures):
            result = future.result()
            if result:
                for k, v in result.items():
                    metrics[k].append(v)

    # mean metrics
    return {m: np.mean(vals) for m, vals in metrics.items()}
```

## ML-Based RecSys

### Feature Engineering & Dataset Preparation

```python
positive_edges = random.sample(list(G.edges()), 50000)

nodes = list(G.nodes())
negative_edges = []
while len(negative_edges) < 50000:
    u, v = random.sample(nodes, 2)
    if not G.has_edge(u, v):
        negative_edges.append((u, v))

edges = positive_edges + negative_edges

# -----------------------------
# 3. PRECOMPUTE NODE FEATURES (FIX)
# -----------------------------
degree = dict(G.degree())
in_degree = dict(G.in_degree())
out_degree = dict(G.out_degree())
cluster = nx.clustering(G.to_undirected())
pagerank = nx.pagerank(G, alpha=0.85)

# -----------------------------
# 4. Compute pair-level features
# -----------------------------
def compute_pair_features(G, edges):
    data = []
    UG = G.to_undirected()

    # Precompute fast pair metrics
    jaccard = { (u,v): s for u,v,s in nx.jaccard_coefficient(UG, edges) }
    pa = { (u,v): s for u,v,s in nx.preferential_attachment(UG, edges) }

    for u, v in edges:
        feat = {
            "u": u,
            "v": v,
            "label": 1 if G.has_edge(u, v) else 0,

            # Node-level features
            "degree_u": degree[u],
            "degree_v": degree[v],
            "in_degree_u": in_degree[u],
            "in_degree_v": in_degree[v],
            "out_degree_u": out_degree[u],
            "out_degree_v": out_degree[v],
            "clustering_u": cluster[u],
            "clustering_v": cluster[v],
            "pagerank_u": pagerank[u],
            "pagerank_v": pagerank[v],

            # Pair-level features
            "common_neighbors": len(set(G.successors(u)) & set(G.successors(v))),
            "jaccard": jaccard.get((u,v), 0),
            "preferential_attachment": pa.get((u,v), 0)
        }
        data.append(feat)

    return data

# -----------------------------
# 5. Run feature generation
# -----------------------------
features_df = compute_pair_features(G, edges)
print("Pair-level features completed")
```

### LightGBM Ranker

```python
params = {
    # ---- Core ranking objective ----
    'objective': 'lambdarank',
    'metric': 'ndcg',
    'ndcg_eval_at': [5, 10, 20],     # evaluate NDCG at specific cutoffs

    # ---- Tree structure ----
    'num_leaves': 128,               # 64 → 128 for more expressiveness
    'max_depth': -1,
    'min_data_in_leaf': 20,          # avoids overfitting groups
    'min_sum_hessian_in_leaf': 1e-3,

    # ---- Boosting ----
    'boosting_type': 'gbdt',
    'n_estimators': 500,             # you can remove if using num_boost_round
    'learning_rate': 0.01,           # LR often higher for ranking
    'num_boost_round': 2000,         # ranking benefits from many rounds

    # ---- Regularization ----
    'lambda_l1': 0.2,
    'lambda_l2': 0.5,
    'min_gain_to_split': 0.0,
    'max_bin': 255,

    # ---- Sampling (important for large imbalanced ranking sets) ----
    'bagging_fraction': 0.9,
    'bagging_freq': 1,
    'feature_fraction': 0.9,

    # ---- Special ranking parameters ----
    'label_gain': [0, 1],            # gains for labels (can do multi-rel)
    'lambdarank_truncation_level': 20,  # reduces noise in pairwise sampling
    'max_position': 20,              # focuses ranking on top-k positions

    # ---- Optimization ----
    'verbosity': -1,
}
```

```python
clf = lgb.LGBMRanker(**params)


clf.fit(
    X_train, y_train,
    group=group_train,
    eval_set=[(X_val, y_val)],
    eval_group=[group_val],
    eval_metric='ndcg'
)
```

### Evaluation

```python
import numpy as np
import math

# ------------------------------
# 1. Ranking metric functions
# ------------------------------

def precision_at_k(y_true, y_scores, k=10):
    y_true = np.array(y_true)
    y_scores = np.array(y_scores)
    idx = np.argsort(y_scores)[::-1][:k]
    top_k_true = y_true[idx]
    return top_k_true.sum() / k

def recall_at_k(y_true, y_scores, k=10):
    y_true = np.array(y_true)
    y_scores = np.array(y_scores)
    idx = np.argsort(y_scores)[::-1][:k]
    top_k_true = y_true[idx]
    return top_k_true.sum() / np.sum(y_true) if np.sum(y_true) > 0 else 0

def mrr_at_k(y_true, y_scores, k=10):
    y_true = np.array(y_true)
    y_scores = np.array(y_scores)
    idx = np.argsort(y_scores)[::-1][:k]
    for rank, i in enumerate(idx, start=1):
        if y_true[i] == 1:
            return 1 / rank
    return 0

def average_precision_at_k(y_true, y_scores, k=10):
    y_true = np.array(y_true)
    y_scores = np.array(y_scores)
    idx = np.argsort(y_scores)[::-1][:k]
    hits, score = 0, 0
    for i, j in enumerate(idx, start=1):
        if y_true[j] == 1:
            hits += 1
            score += hits / i
    return score / np.sum(y_true) if np.sum(y_true) > 0 else 0

def ndcg_at_k(y_true, y_scores, k=10):
    y_true = np.array(y_true)
    y_scores = np.array(y_scores)
    idx = np.argsort(y_scores)[::-1][:k]
    dcg = 0.0
    for i, j in enumerate(idx, start=1):
        if y_true[j] == 1:
            dcg += 1 / math.log2(i + 1)
    idcg = sum(1 / math.log2(i + 1) for i in range(1, min(np.sum(y_true), k) + 1))
    return dcg / idcg if idcg > 0 else 0

# ------------------------------
# 2. Evaluate per group
# ------------------------------

def evaluate_lgbm_ranker(X, y_true, y_scores, group, k=10):
    """
    X, y_true: validation data
    y_scores: predicted ranking scores
    group: array with number of items per query
    """
    start = 0
    metrics = {'precision': [], 'recall': [], 'mrr': [], 'map': [], 'ndcg': []}

    for g in group:
        end = start + g
        y_true_group = y_true[start:end]
        y_scores_group = y_scores[start:end]

        metrics['precision'].append(precision_at_k(y_true_group, y_scores_group, k))
        metrics['recall'].append(recall_at_k(y_true_group, y_scores_group, k))
        metrics['mrr'].append(mrr_at_k(y_true_group, y_scores_group, k))
        metrics['map'].append(average_precision_at_k(y_true_group, y_scores_group, k))
        metrics['ndcg'].append(ndcg_at_k(y_true_group, y_scores_group, k))

        start = end

    # Average across all queries
    return {m: np.mean(vals) for m, vals in metrics.items()}

# ------------------------------
# 3. Apply evaluation
# ------------------------------

# predict scores
y_scores = clf.predict(X_val)

metrics_at_10 = evaluate_lgbm_ranker(X_val, y_val, y_scores, group_val, k=10)

print("LightGBM Ranking Metrics @10:")
for metric, value in metrics_at_10.items():
    print(f"{metric}: {value:.4f}")
```

For Precision@10 metric, both methods give the similar result. Both PageRank and LightGBM algorithm predict very few of the top 10 recommended items as relevant. 

Also in Recall@10 metric both methods show similar performance but PageRank Algorithm is slightly better than LightGBM ranker. It means PageRank covers the true items at top10 recommended items better than LightGBM.

According to MRR@10, LightGBM ranks the relevant items at top 10 much higher than PageRank. It means LightGBM is better than PageRank in ranking. PageRank algorithm can find the items (because it has higher recall) but cannot rank them very well.

MAP measures both whether the relevant items are found and how early they appear in the ranking. Higher MAP means: relevant items are found and appear near the top. According to these results, LightGBM can capture the relevant items and rank them accordingly.

NDCG also proves that LighGBM much more succesfull than PageRank regarding to ranking performance.
